```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import neighbors
from sklearn.decomposition import PCA
from yellowbrick import neighbors as nb

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target
```

```python
model = neighbors.KNeighborsClassifier(10)
visualizer = nb.KnnDecisionBoundariesVisualizer(model, classes=iris.target_names, features=iris.feature_names[:2])
visualizer.fit_draw_poof(X, y)
    
```
